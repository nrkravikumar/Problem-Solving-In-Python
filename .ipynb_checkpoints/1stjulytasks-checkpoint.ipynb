```python
import pandas as pd
dfp='DataFiles/Income.csv'
def readcsvFile(dfp):
    return pd.read_csv(dfp)
df=readcsvFile(dfp)
df
```

```python
def getrowkeyindex(df,row_key):
    values=df.values
    for row in range(len(values)):
#         print(values[row][0],values[row][1])
        if row_key == values[row][0] or row_key == values[row][1]:
            return row
    return
print(getrowkeyindex(df,'Alabama'))  
def getcolkeyindex(df,col_key):
    columns=df.columns
    for c in range(len(columns)):
        if columns[c] == col_key:
            return c
    return
print(getcolkeyindex(df,'2005')) 
def printsinglerow(df):
    x=getrowkeyindex(df,'Alabama')
    for row in range(len(df.values)):
        if row == x:
            print(df.values[row])
    return
printsinglerow(df)
```
