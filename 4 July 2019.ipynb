```python
# Linear graph

import matplotlib.pyplot as plt
plt.title('Basic Line Graph')
plt.plot([1,2,3],[1,2,3])
plt.xlabel("Per")
plt.ylabel("Marks")
plt.show()
```

```python
#Graph to plot distance and time

distance = list(range(2,11,2))
time = list(range(1,6))
# d= 2 * t
plt.plot(time,distance,'bo')
plt.title("Distance Vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid()
plt.show()
```

```python
# v = a t
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i] * i for i in range(len(t))]
plt.plot(t,v,'go')
plt.grid()
plt.show()
```

```python
# Numbers Vs Cubes
n = list(range(10))
sqn = [i ** 3 for i in n]
plt.title("Numbers Vs Cubes")
plt.xlabel("Numbers")
plt.ylabel("Cubes")
plt.plot(n,sqn)
plt.show()
```

```python
# Voltage - Time
import numpy as np
# t = [i/10 for i in range(1,21)]
t = np.arange(0.1,2.0,0.1)
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.title("Time Vs Voltge Graph")
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Graph.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013

import pandas as pd

df=pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]
incomeArkansas = df.values[3,2:]
plt.plot(years,incomeArkansas)
plt.xlabel('Years')
plt.ylabel('IncomeArkansas')
plt.savefig('Images/incomearkansas.png')
plt.show()
```

```python
#Income of all states from 2005 to 2013
df
incomeofallstates = []

for i in df.columns[2:]:
    incomeofallstates.append(df[i].mean())

plt.title('Income Report for all States')
plt.plot(years,incomeofallstates)
plt.xlabel('Years')
plt.ylabel('Income of all States')
plt.grid()
plt.show()
```
